## Implement Flowise Integration

### Subtask:
Write a Python function that sends a user query to the Flowise API, handles authentication (if necessary), and processes the response to extract the search results.


**Reasoning**:
To implement the Flowise integration, I need to define the `flowise_search` function that will make an API call to the Flowise endpoint using the `requests` library, handle the payload and authentication, and extract the relevant response.



```python
import requests
import os

# Retrieve API URL and Key from kernel variables
FLOWISE_API_URL = os.environ.get("FLOWISE_API_URL", "https://cloud.flowiseai.com/api/v1/prediction/eb1731c3-6331-402b-8139-693845e2aeda")
#FLOWISE_API_KEY = os.environ.get("FLOWISE_API_KEY", "YOUR_FLOWISE_API_KEY_HERE") # Placeholder if not set as env var

def flowise_search(user_query):
    """Sends a user query to the Flowise API and returns the search result."""
    payload = {"question": user_query}
    headers = {}

    # Add API key for authentication if it's not the placeholder
    if FLOWISE_API_KEY and FLOWISE_API_KEY != "YOUR_FLOWISE_API_KEY_HERE":
        headers["Authorization"] = f"Bearer {FLOWISE_API_KEY}"

    try:
        response = requests.post(FLOWISE_API_URL, json=payload, headers=headers)
        response.raise_for_status()  # Raise an exception for HTTP errors (4xx or 5xx)

        response_json = response.json()

        # Assuming the search result is under the 'text' key
        if "text" in response_json:
            return response_json["text"]
        else:
            return f"Error: 'text' key not found in Flowise response. Response: {response_json}"

    except requests.exceptions.HTTPError as err:
        return f"HTTP Error: {err}"
    except requests.exceptions.ConnectionError as err:
        return f"Connection Error: {err}"
    except requests.exceptions.Timeout as err:
        return f"Timeout Error: {err}"
    except requests.exceptions.RequestException as err:
        return f"An unexpected error occurred: {err}"
    except ValueError: # JSONDecodeError for older requests versions
        return f"Error: Could not decode JSON response from Flowise. Response content: {response.text}"

print("Flowise search function defined.")
```

**Reasoning**:
Now that the `flowise_search` function is defined, I need to integrate it into the Gradio interface. This involves updating the `gr.Interface` definition to use `flowise_search` as its `fn` (function) and then relaunching the Gradio application.



```python
import gradio as gr
import requests
import os

# Retrieve API URL and Key from kernel variables (already defined in previous cell)
FLOWISE_API_URL = os.environ.get("FLOWISE_API_URL", "https://cloud.flowiseai.com/api/v1/prediction/eb1731c3-6331-402b-8139-693845e2aeda")
#FLOWISE_API_KEY = os.environ.get("FLOWISE_API_KEY", "YOUR_FLOWISE_API_KEY_HERE")

def flowise_search(user_query):
    """Sends a user query to the Flowise API and returns the search result."""
    payload = {"question": user_query}
    headers = {}

    # Add API key for authentication if it's not the placeholder
    if FLOWISE_API_KEY and FLOWISE_API_KEY != "YOUR_FLOWISE_API_KEY_HERE":
        headers["Authorization"] = f"Bearer {FLOWISE_API_KEY}"

    try:
        response = requests.post(FLOWISE_API_URL, json=payload, headers=headers)
        response.raise_for_status()  # Raise an exception for HTTP errors (4xx or 5xx)

        response_json = response.json()

        # Assuming the search result is under the 'text' key
        if "text" in response_json:
            return response_json["text"]
        else:
            return f"Error: 'text' key not found in Flowise response. Response: {response_json}"

    except requests.exceptions.HTTPError as err:
        return f"HTTP Error: {err}"
    except requests.exceptions.ConnectionError as err:
        return f"Connection Error: {err}"
    except requests.exceptions.Timeout as err:
        return f"Timeout Error: {err}"
    except requests.exceptions.RequestException as err:
        return f"An unexpected error occurred: {err}"
    except ValueError: # JSONDecodeError for older requests versions
        return f"Error: Could not decode JSON response from Flowise. Response content: {response.text}"


# Create the Gradio interface with the integrated flowise_search function
iface = gr.Interface(
    fn=flowise_search,
    inputs=gr.Textbox(lines=5, placeholder="Enter your query here..."),
    outputs=gr.Textbox(label="Search Results", lines=10),
    title="Security document search",
    description="Enter your query to search documents via Flowise."
)

print("Gradio interface updated with Flowise integration.")
iface.launch()
```

```python
import gradio as gr
import requests

# üß† Replace with your Flowise API endpoint
FLOWISE_API_URL = "https://cloud.flowiseai.com/api/v1/prediction/eb1731c3-6331-402b-8139-693845e2aeda"

def chat_with_flowise(message, history):
    """
    Sends user message and chat history to Flowise API, returns Flowise response.
    """
    try:
        # Format the conversation history for Flowise
        payload = {
            "question": message,
            "history": history
        }

        # Send the request to Flowise
        response = requests.post(FLOWISE_API_URL, json=payload)
        response.raise_for_status()
        result = response.json()

        # Extract Flowise's text output
        answer = result.get("text", "‚ö†Ô∏è No response received from Flowise.")
        history.append((message, answer))
        return history, ""

    except Exception as e:
        history.append((message, f"‚ùå Error: {str(e)}"))
        return history, ""

# üß© Build the Gradio UI
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("## üíº Security Document Assistant (Powered by Flowise)")
    chatbot = gr.Chatbot(height=400)

    with gr.Row():
        msg = gr.Textbox(
            placeholder="Type your question here (e.g., 'List encryption standards in the policy')...",
            show_label=False,
            scale=4
        )
        send = gr.Button("üöÄ Send", scale=1)
        clear = gr.Button("üßπ Clear", scale=1)

    # üîÅ Add interactivity
    send.click(chat_with_flowise, [msg, chatbot], [chatbot, msg])  # Send button
    msg.submit(chat_with_flowise, [msg, chatbot], [chatbot, msg])  # Enter key
    clear.click(lambda: None, None, chatbot, queue=False)           # Clear chat

# üöÄ Launch the app
demo.launch()
```
